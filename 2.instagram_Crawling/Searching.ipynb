```python
# 크롤링 데이터 불러오기
import pandas as pd
raw_total = pd.read_excel('./files/1_crawling_daegucafe_500.xlsx')
raw_total.head()
```

```python
# 추출 단어 포함하는 데이터 선택하기
select_word = "수성못"

check_list =[]
for content in  raw_total['content']:
    if select_word in content:
        check_list.append(True)
    else:
        check_list.append(False)

select_df = raw_total[check_list]
select_df.head()
```

```python
# 선택 데이터 확인하기
for i in select_df.index:
    print(select_df.loc[i,'content'])
    print('-'*50)
```

```python
# 선택 데이터 저장하기
fpath = f'./files/selected_data_{select_word}.xlsx'
select_df.to_excel(fpath)
```

```python
# 여러개의 단어 선택/추출/저장하기
select_word_list = ['수성못','앞산','교동','삼덕동']
for select_word in select_word_list:
    check_list = []
    for content in raw_total['content']:
        if select_word in content:
            check_list.append(True)
        else:
            check_list.append(False)
    select_df = raw_total[check_list]
    fpath =  f'./files/selected_data_{select_word}.xlsx'
    select_df.to_excel(fpath)
```
