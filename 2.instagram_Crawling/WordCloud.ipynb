```python
# 크롤링 결과 중 해시태그 데이터 불러오기
import pandas as pd
raw_total = pd.read_excel('./files/1_crawling_daegucafe_500.xlsx')
raw_total['tags'][:10]
```

```python
# 해시태그 통합 저장하기
tags_total = []

for tags in raw_total['tags']:
    tags_list = tags[2:-2].split("', '")
    for tag in tags_list:
        tags_total.append(tag)
        
tags_total
```

```python
# 빈도수 집계하기
from collections import Counter
tag_counts = Counter(tags_total)
```

```python
# 가장 많이 사용된 해시태그 살펴보기
tag_counts.most_common(60)
```

```python
# 데이터 정제하기
StopWords = ['','#일상','#맞팔','#ootd','#좋아요','#좋아요반사','#좋반','#데일리룩','#팔로우','#선팔','#데일리','#daily','#오오티디',
            '#소통','#선팔하면맞팔']
tag_total_selected = []
for tag in tags_total:
    if tag not in StopWords:
        tag_total_selected.append(tag)
        
tags_counts_selected = Counter(tag_total_selected)
tags_counts_selected.most_common(50)
```

```python
# 시각화  라이브러리 호출 -> 막대 차트로 해시태그 살펴보기
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc
import sys

if sys.platform in ['win32', 'win64']:
    font_name = 'malgun gothic'
elif sys.platform == 'darwin':
    font_name = 'AppleGothic'
    
rc('font',family = font_name)
```

```python
# 데이터 준비하기
tag_counts_df = pd.DataFrame(tags_counts_selected.most_common(30)) 
tag_counts_df.columns = ['tags','counts']
```

```python
# 막대 차트 그리기
plt.figure(figsize = (10,8))
sns.barplot(x='counts',y='tags',data = tag_counts_df)
plt.show()
```

```python
# 워드클라우드 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import platform 

if platform.system() == 'Windows':
    font_path = "c:/Windows/Fonts/malgun.ttf"
elif platform.system() == 'Darwom':
    font_path = "/User/$USER/Library/Fonts/AppleGothic.ttf"
```

```python
# 워드클라우드 만들기
wordcloud = WordCloud(font_path= font_path,background_color ='white',max_words=100,relative_scaling=0.3,width= 800,
                     height=400).generate_from_frequencies(tags_counts_selected)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('./files/tag_wordcloud.png')
plt.show()
```
