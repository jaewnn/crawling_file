```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
driver = webdriver.Chrome('C:\\Users\82104\python level2\chromedriver.exe')
```

```python
driver.get("https://www.youtube.com/watch?v=4TWR90KJl84")
driver.implicitly_wait(time_to_wait=2)
```

```python
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
last_page_height
```

```python
# Ïä§ÌÅ¨Î°§ÏùÄ Î™áÎ≤à ÎÇ¥Î†§ÏÑú ÎåìÍ∏ÄÏùÑ Í∞ÄÏ†∏Ïò¨Íπå
for i in range(15):
    driver.execute_script(("window.scrollTo(0, document.documentElement.scrollHeight)"))
    time.sleep(3)
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
```

```python
html_source = driver.page_source
```

```python
soup = BeautifulSoup(html_source,'html.parser')
```

```python
youtube_user_ID = soup.select('#author-text > span')
youtube_comment = soup.select('#content-text')[1].text.strip()
print(youtube_user_ID,'\n',youtube_comment)
```

```python
youtube_user_IDs = []
youtube_comments = []

for i in range(len(youtube_user_ID)):
    str_tmp = str(soup.select('#author-text > span')[i].text.strip())
    print(str_tmp)
    youtube_user_IDs.append(str_tmp)
    
    str_tmp = str(soup.select('#content-text')[i].text.strip())
    print(str_tmp)
    youtube_comments.append(str_tmp)
```

```python
import pandas as pd 
pd_data = {"ID":youtube_user_IDs,"Comments":youtube_comments}
youtube_df = pd.DataFrame(pd_data)
```
