```python
playlists = youtube.playlists().list(
    channelId = channel_id,
    part = "snippet",
    maxResults = 20
).execute()
```

```python
import pandas as pd
playlist_titles = []
playlist_ids = []
```

```python
for i in playlists['items']:
    playlist_titles.append(i['snippet']['title'])
    playlist_ids.append(i['id'])
```

```python
playlist_df = pd.DataFrame([playlist_titles,playlist_ids]).T
playlist_df.columns = ['playlist','playlist_id']
```

```python
playlist_df
```

```python
video_ids =[]
video_titles = []
video_dates =[]

for playlist_id in playlist_df['playlist_id']:
    playlist_videos = youtube.playlistItems().list(
        playlistId = playlist_id,
        part = 'snippet',
        maxResults = 50
    )
    while playlist_videos:
        playlistitems_list_response = playlist_videos.execute()
        
        for playlist_item in playlistitems_list_response['items']:
            title = playlist_item['snippet']['title']
            video_id = playlist_item['snippet']['resourceId']['videoId']
            video_ids.append(video_id)
            video_titles.append(title)
        
        playlist_videos = youtube.playlistItems().list_next(
            playlist_videos,playlistitems_list_response
        )
```

```python
video_df = pd.DataFrame()
video_df['Title'] = video_titles
video_df['Video Id'] = video_ids
video_df
```
